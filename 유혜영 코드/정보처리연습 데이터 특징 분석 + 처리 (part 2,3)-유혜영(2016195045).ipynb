## 2. Concatenate the five dataframes into one dataframe and check for null data in the dataset

```python
# Group the dataframes into one list
grouped_frames = [expedia_org_df1, booking_org_df1, agoda_org_df1, priceline_org_df1, hostel_org_df1]
# Concatenate the dataframes on grouped_frames and save it to a variable called grouped_df 
grouped_df = pd.concat(grouped_frames)
```

## 3. Put the company_name column at the end of dataframe 

```python
# Put the company_name column at the end
preprocessed_df = grouped_df[['review_numbers', 'review_scores', 'review_weekdays', 'sentiment_compound', 'company_name']]
```

```python
# Check if the company_name column has been placed at the last column
preprocessed_df.head(5)
```

## 4. Change the categorical values to numerical values

```python
# Check which data types are in the dataframe
preprocessed_df.info()
```

```python
# Select only the columns that contain object values
preprocessed_df_cat_columns = preprocessed_df.select_dtypes(['object']).columns[:-1]
```

```python
# Check the columns
preprocessed_df_cat_columns
```

```python
# Use loop to change the object type to category type
for i in preprocessed_df_cat_columns:
    preprocessed_df[i] = preprocessed_df[i].astype('category')
```

```python
# Check the change from object type to category type
preprocessed_df.info()
```

```python
# Change the categorical value to numerical value using lambda and apply
preprocessed_df[preprocessed_df_cat_columns] = preprocessed_df[preprocessed_df_cat_columns].apply(lambda x: x.cat.codes)
```

```python
# Check the change
preprocessed_df.head(5)
```

## (2) Value Count Tables for Concatenated Dataframe

```python
# Import pandas.plotting.table
from pandas.plotting import table
# Examine weekday value counts for the concatenated dataframe
all_weekday_vc_df = grouped_df1['review_weekdays'].value_counts().to_frame('Concatenated Dataframe Weekday Value Counts')
ax = plt.subplot(frame_on=False) # no visible frame
ax.xaxis.set_visible(False)  # hide the x axis
ax.yaxis.set_visible(False)  # hide the y axis
table1 = table(ax, all_weekday_vc_df, loc='upper right', colWidths=[0.95]*len(all_weekday_vc_df.columns))
table1.auto_set_font_size(False) # Activate set fontsize manually
table1.set_fontsize(12) # if ++fontsize is necessary ++colWidths
plt.savefig('/Users/angieryu2202/Desktop/all_weekday_vc_df.png', bbox_inches='tight', transparent=True)
```

```python
# Examine valence value counts for the concatenated dataframe
all_valence_vc_df = grouped_df1['sentiment_compound'].value_counts().to_frame('Concatenated Dataframe Valence Value Counts')
ax = plt.subplot(frame_on=False) # no visible frame
ax.xaxis.set_visible(False)  # hide the x axis
ax.yaxis.set_visible(False)  # hide the y axis
table1 = table(ax, all_valence_vc_df, loc='upper right', colWidths=[0.95]*len(all_weekday_vc_df.columns))
table1.auto_set_font_size(False) # Activate set fontsize manually
table1.set_fontsize(12) # if ++fontsize is necessary ++colWidths
plt.savefig('/Users/angieryu2202/Desktop/all_valence_vc_df.png', bbox_inches='tight', transparent=True)
```
